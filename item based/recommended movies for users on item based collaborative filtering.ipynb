```python
import numpy as np
import pandas as pd
from math import sqrt
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from sklearn.metrics import pairwise_distances
from sklearn.neighbors import NearestNeighbors
from scipy.stats import pearsonr
```

```python
movies=pd.read_csv("movies.csv")
ratings=pd.read_csv("ratings.csv")
movies.head(1)
```

```python
## keping all those users which has rated a minimum of 10 movies
v = ratings.userId.value_counts()
v_df=ratings[ratings.userId.isin(v.index[v.gt(10)])]
v
```

```python
## taking user id 
userid=320
```

```python
df=v_df.loc[ratings['userId']==userid]
sdf=df.sort_values('rating',ascending=False)
sdf.head(20)
```

```python
## extracting user top 5 movies and removing all other movies which have been rated by user
x_df=sdf[0:5]
x_df.head()
xdf=pd.merge(x_df,movies,on='movieId').drop(['genres','timestamp'],axis=1)
xdf.head()
```

```python
z_df=pd.concat([sdf,x_df,x_df]).drop_duplicates(keep=False)
z_df.drop(['userId','rating'],axis=1,inplace=True)
```

```python
movies=pd.concat([movies,z_df,z_df]).drop_duplicates(keep=False)
movies.shape
```

```python
new_df=pd.merge(ratings,movies,on='movieId').drop(['genres','timestamp_x','timestamp_y'],axis=1)
new_df.head()
```

```python
## creating movie matrix
movie_matrix=new_df.pivot_table(index='userId',columns='title',values='rating').fillna(0)
```

```python
## funtion for checking movies one by one with user rating
```

```python
def rec(movie,user_rat):
    df=pd.DataFrame()
    movie_user_rating=movie_matrix[movie]
    similar_movies=movie_matrix.corrwith(movie_user_rating)
    corr_movie=pd.DataFrame(similar_movies,columns=['correlation'])
    corr_movie.dropna(inplace=True)
    df=df.append(corr_movie.sort_values(by='correlation',ascending=False)[1:5])
    df["combined"]=df["correlation"]*user_rat
    return df
```

```python
## funtion using loop for dataframe xdf where user's top rated movies are present with title and ratings
## using the concept of item based collaborative filtering similar movies and correlation first we calculated top four movie
## based on correlation and after that multiplying by user rating for that particular movie we ha calculated combined rating. 
```

```python
def rec2(xdf):
    ndf=pd.DataFrame()
    for ind,row in xdf.iterrows():
        movie_user_rating=movie_matrix[xdf.loc[ind,'title']]
        similar_movies=movie_matrix.corrwith(movie_user_rating)
        corr_movie=pd.DataFrame(similar_movies,columns=['correlation'])
        corr_movie.dropna(inplace=True)
        ndf=ndf.append(corr_movie.sort_values(by='correlation',ascending=False)[1:5])
        ndf['combined']=ndf['correlation']*(xdf.loc[ind,'rating'])
        
    return ndf 
```

```python
gdf=rec2(xdf)
gdf
```

```python
## top five movies recommender for the entered user id
gdf=gdf.sort_values(by='combined',ascending=False)
gdf.head(5)
```
